```python
import xsearch as xs
```

# Create pathDict with initial filtering

```python
pathDict = xs.findPaths('amip', 'pr', 'day', cmipTable='day')
dpaths = sorted(list(pathDict.keys()))
print(len(dpaths))
for path in dpaths:
    print(path)
    
```

# Additional iterative filtering with getValuesForFacet

```python
print('Filtering with mip_era')
pathDict_filt = xs.getValuesForFacet(pathDict, 'mip_era', 'CMIP5')
print(len(pathDict_filt))
for path in pathDict_filt.keys():
    print(path)

print()
print('Filtering with institute')
pathDict_filt = xs.getValuesForFacet(pathDict_filt, 'institute', 'BCC')
print(len(pathDict_filt))
for path in pathDict_filt.keys():
    print(path)
    
```
